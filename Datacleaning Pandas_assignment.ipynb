```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df
```

## Question 1

```python

l=[]
def fill_na(x):
    l.append(x)
    if np.isnan(x):
        return int(l[l.index(x)-1]+10)
    else:
        return int(x)
df['FlightNumber']=df['FlightNumber'].apply(fill_na)
```

```python
df
```

## Question 2

```python


l=list(map(lambda x:x.split('_'),list(df['From_To'])))
df1=pd.DataFrame(l,columns=['From','To'])
```

```python
df1
```

## Question 3

```python

df1['From']=df1['From'].apply(lambda x:x.title())

df1['To']=df1['To'].apply(lambda x:x.title())
```

```python
df1
```

## Question 4

```python

df.drop('From_To',axis=1,inplace=True)
```

```python
new_df=df1.merge(df,left_index=True,right_index=True)
```
